```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('data_v5.xlsx')
```

```python
ban_tweet = pd.read_excel('dengue_bangla.xlsx')
eng_tweet = pd.read_excel('english_tweets.xlsx')
```

```python
ban_tweet = ban_tweet[['datetime','location']]
eng_tweet = eng_tweet[['datetime','location']]
```

```python
data.head(20)
```

```python
data.set_index('locations_month',inplace=True)
data['date'] = ['20'+data.index[i].split('20')[1] for i in range(len(data))]
```

```python
tweet_data = pd.concat([ban_tweet,eng_tweet])
```

```python
tweet_data
```

```python
tweet_data = tweet_data.sample(frac=1).reset_index(drop=True)
```

```python
tweet_data['location'].fillna("0",inplace=True)
```

```python
loc_lst = list(tweet_data['location'])
```

```python
for i in range(len(loc_lst)):
    if ' ' in loc_lst[i]:
        loc_lst[i] = loc_lst[i].replace(' ','')
for i in range(len(loc_lst)):
    if ' ' in loc_lst[i]:
        loc_lst[i] = loc_lst[i].replace(' ','')
for i in range(len(loc_lst)):
    if ' ' in loc_lst[i]:
        loc_lst[i] = loc_lst[i].replace(' ','')
for i in range(len(loc_lst)):
    if 'medical' in loc_lst[i]:
        loc_lst[i] = loc_lst[i].replace('medical','')
for i in range(len(loc_lst)):
    if 'hospital' in loc_lst[i]:
        loc_lst[i] = loc_lst[i].replace('hospital','')
```

```python
loc_lst
```

```python
for i in range(len(loc_lst)):
    if '00' in loc_lst[i]:
        loc_lst[i] = loc_lst[i].replace('00','0')
for i in range(len(loc_lst)):
    if ';' in loc_lst[i] and (len(loc_lst[i]) > 3):
        loc_lst[i] = loc_lst[i].split(';')
    elif ';' in loc_lst[i]:
        loc_lst[i] = loc_lst[i].replace(';','0')
    else:
        loc_lst[i] = [loc_lst[i]]
for i in range(len(loc_lst)):
    if type(loc_lst[i]) != list:
        loc_lst[i] = [loc_lst[i]]
for i in range(len(loc_lst)):
    if '' in loc_lst[i]:
        loc_lst[i].remove('')
for i in range(len(loc_lst)):
    if '' in loc_lst[i]:
        loc_lst[i].remove('')
for i in range(len(loc_lst)):
    if '' in loc_lst[i]:
        loc_lst[i].remove('')        
```

```python
tweet_data['year_month'] = [str(np.array(tweet_data['datetime'])[i]).split('T')[0] for i in range(len(tweet_data))]
```

```python
tweet_data.head(500)
```

```python
temp_date = []
for i in range(len(tweet_data)):
    temp2 = str(np.array(tweet_data['year_month'])[i]).split('-')[1]
    if '10' in temp2 :
        temp2 = temp2.replace('10','10')
    elif '0' in temp2:
        temp2 = temp2.replace('0','')
    temp1 = str(np.array(tweet_data['year_month'])[i]).split('-')[0]
    temp_date.append(temp1+'-'+temp2)
```

```python
tweet_data['year_month'] = temp_date
```
